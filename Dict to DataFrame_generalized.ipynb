# Import Libraries

```python
import pandas as pd
import numpy as np
import datetime
```

## Date & Time Functions

```python
# def occurence_time(row):
#     if row['OccurTime'] < 800:
#         return "Morning"
#     elif row['OccurTime'] < 1600:
#         return "Day"
#     else:
#         return "Evening"

def print_month(row):
    cur_date = row['Date']
    year, month, day = (int(x) for x in cur_date.split('-'))    
    month = int(month)
    return month

def day_of_month(row):
    cur_date = row['Date']
    year, month, day = (int(x) for x in cur_date.split('-'))    
    day = int(day)
    return day

def year(row):
    cur_date = row['Date']
    year, month, day = (int(x) for x in cur_date.split('-'))    
    year = int(year)
    return year

def day_of_week(row):
    cur_date = row['Date']
    year, month, day = (int(x) for x in cur_date.split('-'))    
    dt = datetime.date(year, month, day)
    return(dt.weekday())
```

# Manipulate and Export

```python
file = open('generalized.txt','r').read()
user_dict = eval(file)

data = pd.DataFrame.from_dict(user_dict,orient='index')
data.reset_index(inplace=True)

#rename features
print(data.columns)
data.rename(columns={'index': 'Date', 0: 'Category 1', 1: 'Category 2',
                     2: 'Category 3',3: 'Category 4'}, inplace=True)

#Extract date features
data['Year'] = data.apply(lambda row: year(row), axis=1)
data['Month'] = data.apply(lambda row: print_month(row), axis=1)
data['Day of Month'] = data.apply(lambda row: day_of_month(row), axis=1)
data['Day of Week'] = data.apply(lambda row: day_of_week(row), axis=1)
#data['Shift Occurence'] = data.apply (lambda row: occurence_time(row),axis=1)

#Reorder Columns
data = data[['Date','Year','Month','Day of Month','Day of Week'
             ,'Category 1', 'Category 2', 'Category 3', 'Category 4']]
#Export
data.to_csv("cobra-generalized.csv", index=False)
print(data.columns.values)
data
```
